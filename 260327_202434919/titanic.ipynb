```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
import warnings
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/260327_202434919/titanic.csv'

df = pd.read_csv(path)

df
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])

df.head()
```

```python
X = df.drop(['Survived', 'Pclass'], axis=1)
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

df.isnull().sum()
```

```python
#호출시
models = {
    "Logistic Regression": LogisticRegression(random_state=1000, max_iter=1000),
    "SVM": SVC(random_state=0),
    "Decision Tree": DecisionTreeClassifier(random_state=0),
    "Random Forest": RandomForestClassifier(random_state=0),
    "KNN": KNeighborsClassifier(n_neighbors=5)
}
```

```python
models["Logistic Regression"].fit(X_train, y_train)
pred_lr = models["Logistic Regression"].predict(X_test)
print("\n--- Logistic Regression ---")
print(f"Accuracy: {accuracy_score(y_test, pred_lr):.4f}")
print(confusion_matrix(y_test, pred_lr))
```

```python
# --- SVM ---
models["SVM"].fit(X_train, y_train)
pred_svm = models["SVM"].predict(X_test)
print("\n--- SVM ---")
print(f"Accuracy: {accuracy_score(y_test, pred_svm):.4f}")
print(confusion_matrix(y_test, pred_svm))
```

```python
# --- Decision Tree ---
models["Decision Tree"].fit(X_train, y_train)
pred_dt = models["Decision Tree"].fit(X_train, y_train).predict(X_test)
print("\n--- Decision Tree ---")
print(f"Accuracy: {accuracy_score(y_test, pred_dt):.4f}")
print(confusion_matrix(y_test, pred_dt))
```

```python
# --- Random Forest ---
models["Random Forest"].fit(X_train, y_train)
pred_rf = models["Random Forest"].predict(X_test)
print("\n--- Random Forest ---")
print(f"Accuracy: {accuracy_score(y_test, pred_rf):.4f}")
print(confusion_matrix(y_test, pred_rf))
```

```python
# --- KNN ---
models["KNN"].fit(X_train, y_train)
pred_knn = models["KNN"].predict(X_test)
print("\n--- KNN ---")
print(f"Accuracy: {accuracy_score(y_test, pred_knn):.4f}")
print(confusion_matrix(y_test, pred_knn))
```
