```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
import warnings
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/260320_202434919/abalone.csv'

df = pd.read_csv(path)

X = df.drop(['id', 'Rings', 'Sex'], axis=1)
y = df['Rings']

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf_lr = LogisticRegression(max_iter=1000, random_state=0)
clf_lr.fit(X_train, y_train)
pred_lr = clf_lr.predict(X_test)

print("\n--- 1. Logistic Regression Classifier ---")
print(f"Accuracy: {accuracy_score(y_test, pred_lr):.4f}")
print(confusion_matrix(y_test, pred_lr))
```

```python
clf_svm = SVC(kernel='rbf', random_state=0)
clf_svm.fit(X_train, y_train)
pred_svm = clf_svm.predict(X_test)

print("\n--- 2. SVM Classifier ---")
print(f"Accuracy: {accuracy_score(y_test, pred_svm):.4f}")
print(confusion_matrix(y_test, pred_svm))
```

```python
clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, y_train)
pred_dt = clf_dt.predict(X_test)

print("\n--- 3. Decision Tree Classifier ---")
print(f"Accuracy: {accuracy_score(y_test, pred_dt):.4f}")
print(confusion_matrix(y_test, pred_dt))
```

```python
clf_rf = RandomForestClassifier(n_estimators=100, random_state=0)
clf_rf.fit(X_train, y_train)
pred_rf = clf_rf.predict(X_test)

print("\n--- 4. Random Forest Classifier ---")
print(f"Accuracy: {accuracy_score(y_test, pred_rf):.4f}")
print(confusion_matrix(y_test, pred_rf))
```

```python
# --- KNN ---
clf_knn = KNeighborsClassifier(n_neighbors=5)
clf_knn.fit(X_train, y_train)
pred_knn = clf_knn.predict(X_test)

print("\n--- 5. K-Nearest Neighbors Classifier ---")
print(f"Accuracy: {accuracy_score(y_test, pred_knn):.4f}")
print(confusion_matrix(y_test, pred_knn))
```
